## Step 1: Library Ingest and Data Import

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the master analytical dataset
df = pd.read_csv('../data/master_dataset.csv')
print(f"Master Dataset Dimensions: {df.shape}")
df.head()
```

## Step 2: Preprocessing and Missing Value Diagnostics

```python
# Check for missing values
missing = df.isnull().sum()
print("Missing values per column:")
print(missing[missing > 0])
```

## Step 3: Outlier Profiling (IQR Method)

```python
def plot_outliers(df, column):
    plt.figure(figsize=(10, 4))
    sns.boxplot(x=df[column], palette='Set2')
    plt.title(f'Outlier Boxplot for {column.capitalize()}')
    plt.show()

plot_outliers(df, 'sales_amount')
plot_outliers(df, 'profit')
```

## Step 4: Exploratory Analysis (Monthly Sales Trends & Seasonality)

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df_monthly = df.set_index('order_date').resample('ME')['revenue'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_monthly, x='order_date', y='revenue', marker='o', color='#1a5276', linewidth=2.5)
plt.title('Monthly Sales Trend (Gross Revenue)', fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Revenue ($)')
plt.grid(True)
plt.show()
```

## Step 5: Customer Loyalty RFM Segmentation

```python
df_rfm = pd.read_csv('../data/rfm_customer_segments.csv')
print("RFM Segment Counts:")
print(df_rfm['rfm_segment'].value_counts())

plt.figure(figsize=(10, 5))
sns.countplot(data=df_rfm, y='rfm_segment', order=df_rfm['rfm_segment'].value_counts().index, palette='viridis')
plt.title('Customer Loyalty Segment Distribution (RFM)', fontsize=14, fontweight='bold')
plt.xlabel('Customer Count')
plt.ylabel('Loyalty Segment')
plt.show()
```
